## **Business Problem Understanding**

**Context :**

Syarah, sebuah perusahaan yang bergerak di bidang jual beli mobil bekas, beroperasi dalam pasar yang sangat kompetitif, di mana penetapan harga yang tepat sangat penting untuk menarik pembeli dan mempercepat proses penjualan. Perusahaan ingin mengetahui faktor-faktor yang memengaruhi harga jual, seperti usia, kondisi mobil, dan jarak tempuh, karena sering kali mobil tidak terjual dalam waktu yang diharapkan atau dijual dengan margin keuntungan yang tipis. Akibatnya, terjadi penumpukan stok dan peningkatan biaya operasional, membuat upaya pemasaran dan pengelolaan inventaris menjadi sia-sia.

Agar tetap kompetitif dan meningkatkan efisiensi bisnis, Syarah perlu mengoptimalkan strategi penetapan harga yang tepat melalui analisis mendalam terhadap data yang tersedia. Dengan pemahaman yang lebih baik tentang tren pasar dan faktor yang memengaruhi harga mobil bekas, Syarah dapat mengurangi waktu penjualan, memaksimalkan keuntungan, dan mengurangi pemborosan sumber daya, sehingga meningkatkan efektivitas operasional perusahaan secara keseluruhan.

**Problem Statement :**

**Perusahaan mengalami kesulitan dalam menentukan harga jual mobil bekas yang tepat karena belum memahami faktor-faktor utama yang mempengaruhinya. Akibatnya, harga yang ditetapkan bisa tidak sesuai, menyebabkan mobil tidak terjual dalam waktu lama atau margin keuntungan menurun, serta meningkatkan biaya operasional dan menghambat pendapatan.**

Untuk mengatasi ini, perusahaan perlu menganalisis faktor-faktor seperti usia mobil, kondisi mesin, jarak tempuh, dan lokasi penjualan. Dengan demikian, harga dapat ditetapkan lebih akurat, mempercepat penjualan, dan mengurangi pemborosan sumber daya, sehingga meningkatkan keuntungan dari penjualan mobil bekas.

**Goals :**

Berdasarkan permasalahan tersebut, **Syarah ingin mengetahui harga jual mobil bekas secara akurat untuk mempercepat penjualan dan meningkatkan keuntungan.** Saat ini, perusahaan ingin mengetahui faktor-faktor seperti usia mobil, kondisi mesin, jarak tempuh, dan lokasi penjualan yang membuat harga menjadi relevan dan tepat, sehingga harga yang ditetapkan sesuai dengan kondisi pasar, menghindari mobil sulit terjual, dan menekan biaya operasional yang meningkat.

**Analytic Approach :**

Kita akan menganalisis data mobil bekas untuk menemukan pola yang mempengaruhi harga jual mobil. Kita akan mengeksplorasi faktor-faktor seperti usia mobil, kondisi mesin, jarak tempuh, dan lokasi penjualan yang berkontribusi terhadap penentuan harga.

Kemudian, kita akan membangun model prediksi harga yang dapat membantu Syarah memprediksi harga jual optimal berdasarkan variabel-variabel tersebut. Dengan model ini, perusahaan dapat menetapkan harga mobil bekas yang lebih akurat dan sesuai dengan pasar, mempercepat proses penjualan, dan meningkatkan keuntungan.

**Metric Evaluation :**

Dalam prediksi harga jual mobil bekas, kita akan menggunakan ***Mean Absolute Percentage Error* (MAPE) sebagai metrik evaluasi utama.** MAPE akan membantu kita memahami seberapa akurat model memprediksi harga relatif terhadap harga aktual dalam bentuk persentase, yang sangat relevan dalam konteks penetapan harga. Dengan MAPE, Syarah dapat melihat seberapa jauh prediksi harga meleset secara proporsional, sehingga lebih mudah untuk menilai kesalahan prediksi dalam konteks berbagai kisaran harga mobil bekas.

Selain itu, kita juga akan menggunakan *Mean Absolute Error* (MAE), *Root Mean Square Error* (RMSE), dan R-squared sebagai metrik pembantu. MAE akan memberikan pandangan rata-rata kesalahan prediksi dalam satuan harga yang sebenarnya, membantu perusahaan memahami seberapa jauh prediksi harga meleset dalam nominal rupiah. RMSE akan digunakan untuk memberikan bobot lebih besar pada kesalahan besar, sehingga membantu mendeteksi prediksi yang jauh meleset, yang penting dalam menjaga akurasi tinggi untuk mobil dengan harga yang lebih mahal. Terakhir, koefisien determinasi (R-squared) akan menunjukkan seberapa besar variasi dalam harga jual mobil bekas yang dapat dijelaskan oleh model kita. Nilai R-squared yang mendekati 1 menunjukkan bahwa model kita mampu menjelaskan sebagian besar variasi dalam data, sedangkan nilai yang mendekati 0 menunjukkan bahwa model kita kurang efektif dalam menjelaskan data.

Sekarang hanya tersisa missing value pada Engine_Size yang akan kita isi dengan imputer pada pipeline nantinya. Setelah ini kita akan melihat metode mana yang cocok sebagai imputer pada Engine_Size

#### **Engine_Size**

#### **Median**

```python
# Menyimpan data sebelum imputasi
df_before_imputation = df_cleaned.copy()

# Menghitung median Engine_Size
median_engine_size = df_cleaned['Engine_Size'].median()

# Membuat salinan dataframe untuk simulasi imputasi
df_after_imputation = df_cleaned.copy()
df_after_imputation['Engine_Size'] = df_after_imputation['Engine_Size'].fillna(median_engine_size)

# Membandingkan sebelum dan sesudah imputasi
comparison = pd.DataFrame({
    'Before Imputation': df_before_imputation['Engine_Size'].describe(),
    'After Imputation': df_after_imputation['Engine_Size'].describe()
})

print("Perbandingan statistik Engine_Size sebelum dan sesudah imputasi dengan median:")
display(comparison)

print("\nJumlah missing value pada Engine_Size:")
print(f"Sebelum imputasi: {df_before_imputation['Engine_Size'].isna().sum()}")
print(f"Setelah imputasi: {df_after_imputation['Engine_Size'].isna().sum()}")

# Membuat histogram untuk perbandingan visual
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
df_before_imputation['Engine_Size'].hist(bins=30, edgecolor='black')
plt.title('Distribusi Engine_Size Sebelum Imputasi')
plt.xlabel('Engine_Size')
plt.ylabel('Frekuensi')

plt.subplot(1, 2, 2)
df_after_imputation['Engine_Size'].hist(bins=30, edgecolor='black')
plt.title('Distribusi Engine_Size Setelah Imputasi')
plt.xlabel('Engine_Size')
plt.ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

Evaluasi Imputasi Missing Value dengan Median

Berdasarkan hasil yang ditampilkan sebelumnya, kita dapat menyimpulkan bahwa imputasi missing value pada kolom Engine_Size menggunakan median berhasil dilakukan:

1. Jumlah missing value:
   - Sebelum imputasi: 2141 (berdasarkan selisih jumlah data sebelum dan sesudah imputasi)
   - Setelah imputasi: 0

2. Statistik deskriptif:
   - Mean sedikit berubah dari 3.175498 menjadi 3.050123
   - Standar deviasi menurun dari 1.567450 menjadi 1.361217
   - Nilai minimum, maksimum, dan median (50%) tetap sama

3. Distribusi data:
   - Histogram menunjukkan perubahan distribusi, dengan peningkatan frekuensi di sekitar nilai median (2.7)

Kesimpulan:
- Imputasi berhasil menghilangkan semua missing value
- Distribusi data secara umum tetap terjaga, meskipun ada sedikit perubahan
- Penggunaan median sebagai metode imputasi cukup efektif dalam menjaga karakteristik data asli

Catatan:
Meskipun imputasi berhasil, perlu diingat bahwa metode ini memiliki keterbatasan. Untuk analisis lebih lanjut, mungkin perlu mempertimbangkan metode imputasi yang lebih canggih atau menggunakan teknik lain seperti penanganan missing value dalam model machine learning.

#### **Iterative Imputer**

```python
from sklearn.experimental import enable_iterative_imputer  # Enables IterativeImputer
from sklearn.impute import IterativeImputer

# Menyimpan data sebelum imputasi
df_before_imputation = df_cleaned.copy()

# Membuat objek IterativeImputer
imputer = IterativeImputer()

# Melakukan imputasi pada kolom 'Engine_Size'
df_after_imputation = df_cleaned.copy()
df_after_imputation['Engine_Size'] = imputer.fit_transform(df_after_imputation[['Engine_Size']])


# Membandingkan sebelum dan sesudah imputasi
comparison = pd.DataFrame({
    'Before Imputation': df_before_imputation['Engine_Size'].describe(),
    'After Imputation': df_after_imputation['Engine_Size'].describe()
})

print("Perbandingan statistik Engine_Size sebelum dan sesudah imputasi dengan Iterative Imputer:")
display(comparison)

print("\nJumlah missing value pada Engine_Size:")
print(f"Sebelum imputasi: {df_before_imputation['Engine_Size'].isna().sum()}")
print(f"Setelah imputasi: {df_after_imputation['Engine_Size'].isna().sum()}")

# Membuat histogram untuk perbandingan visual

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
df_before_imputation['Engine_Size'].hist(bins=30, edgecolor='black')
plt.title('Distribusi Engine_Size Sebelum Imputasi')
plt.xlabel('Engine_Size')
plt.ylabel('Frekuensi')

plt.subplot(1, 2, 2)
df_after_imputation['Engine_Size'].hist(bins=30, edgecolor='black')
plt.title('Distribusi Engine_Size Setelah Imputasi')
plt.xlabel('Engine_Size')
plt.ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

Perbandingan Hasil Imputasi: Median Imputer vs Iterative Imputer untuk Pipeline

Berdasarkan hasil imputasi yang telah dilakukan sebelumnya menggunakan metode Median dan Iterative, kedua metode menghasilkan distribusi yang cukup mirip untuk kolom `Engine_Size`. Namun, dalam konteks memasukkan metode imputasi ke dalam pipeline, Iterative Imputer lebih cocok karena:

1. Akurasi yang Lebih Baik:
   - Iterative Imputer mempertimbangkan hubungan antar variabel dalam dataset ketika melakukan imputasi.
   - Ini berarti bahwa model dapat menggunakan informasi dari variabel lain untuk memperkirakan nilai `Engine_Size` yang hilang, yang dapat menghasilkan imputasi yang lebih akurat dibandingkan dengan hanya menggunakan median.
   - Dalam konteks data mobil bekas, variabel-variabel seperti Type, Year, dan Mileage dapat berkorelasi dengan `Engine_Size`. Iterative Imputer dapat memanfaatkan korelasi ini untuk membuat prediksi yang lebih baik daripada hanya mengisi dengan nilai tengah.

2. Menjaga Distribusi Data:
   - Iterative Imputer cenderung mempertahankan bentuk distribusi data asli dibandingkan dengan median imputation.
   - Median imputation dapat mengakibatkan pengurangan variabilitas data, terutama jika jumlah missing value yang besar.
   - Sementara Iterative Imputer berusaha untuk mengisi nilai yang hilang dengan nilai yang konsisten dengan distribusi data yang ada.

3. Mempertimbangkan Hubungan Antar Variabel:
   - Iterative Imputer bekerja dengan mengimputasi nilai yang hilang secara berulang menggunakan model regresi atau lainnya.
   - Model-model ini dapat mempelajari pola dari variabel lain dalam dataset, yang memungkinkan mereka untuk membuat prediksi yang lebih akurat tentang nilai yang hilang.
   - Dalam kasus `Engine_Size`, model dapat mempelajari bahwa mobil dengan type tertentu cenderung memiliki engine size tertentu, atau mobil dengan tahun pembuatan tertentu cenderung memiliki engine size tertentu.

4. Pertimbangan Kompleksitas:
   - Meskipun Iterative Imputer lebih kompleks dan membutuhkan waktu lebih lama untuk dijalankan, dalam konteks analisis dan pemodelan data, akurasi model dan kualitas prediksi lebih diutamakan.
   - Oleh karena itu, trade-off antara waktu komputasi dan akurasi lebih condong ke Iterative Imputer dalam skenario ini.

Dengan mempertimbangkan semua faktor di atas, Iterative Imputer menjadi pilihan yang lebih baik dibandingkan dengan median untuk imputasi `Engine_Size` dalam dataset ini. Hal ini diharapkan dapat memberikan hasil analisis dan pemodelan yang lebih akurat dan relevan.

```python
# Membuat fungsi untuk mengecek outliers
def find_anomalies(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}')
```

```python
# Outliers pada fitur Mileage
find_anomalies(df_cleaned['Mileage'])
print()
# Outliers pada fitur Price
find_anomalies(df_cleaned['Engine_Size'])
print()
# Outliers pada fitur Price
find_anomalies(df_cleaned['Price'])
print()
```

### **Fitur Kategorik**

Kita akan juga melihat hubungan antar kolom menggunakan pendekatan statistik dan dengan domain knowledge.

```python
# Perform one-way ANOVA for categorical columns against Price
from scipy import stats

categorical_columns = ['Make', 'Type', 'Origin', 'Color', 'Options', 'Fuel_Type', 'Gear_Type', 'Region']

for column in categorical_columns:
    groups = [group for _, group in df_cleaned.groupby(column)['Price']]
    _, p_value = stats.f_oneway(*groups)
    print(f"p-value untuk {column}: {p_value}")
    if p_value < 0.05:
        print(f"{column} memiliki pengaruh signifikan terhadap Price (p < 0.05)")
    else:
        print(f"{column} tidak memiliki pengaruh signifikan terhadap Price (p >= 0.05)")
    print("-----------------------------")
```

Berdasarkan Uji Anova, semua memiliki pvalue di bawah 0.05, oleh karena itu kita akan menggunakan semua fitur kategorik yang tersisa.

Selain itu, dari sudut pandang domain knowledge, semua fitur kategorik ini memiliki potensi untuk mempengaruhi harga mobil:
1. Make (Merek): Merek mobil sering kali mempengaruhi harga karena reputasi, kualitas, dan persepsi konsumen.

4. Color (Warna): Beberapa warna mungkin lebih populer dan dapat mempengaruhi harga.
5. Options (Pilihan): Fitur tambahan pada mobil biasanya meningkatkan harganya.
6. Fuel_Type (Jenis Bahan Bakar): Jenis bahan bakar dapat mempengaruhi harga karena perbedaan efisiensi dan biaya operasional.
7. Gear_Type (Jenis Transmisi): Transmisi otomatis umumnya lebih mahal daripada manual.

```python
# Extract General Information (Tipe Data, Missing Value, % Missing Value, Unique, Sample Unique)

list_item = []
for i in df_model.columns :
    list_item.append([
        i,
        df_model[i].dtype,
        df_model[i].nunique(),
        df_model[i].unique()[:9]
    ])
info_df = pd.DataFrame(list_item, columns='Nama Kolom, Tipe Data, Unique, Sample Unique'.split(', '))
info_df
```

Berdasarkan informasi di atas, kita akan mengelompokkan fitur menjadi :

1. Fitur Numerik : `Year`, `Engine_Size`, dan `Mileage`.
   - Karena mereka memang fitur dengan tipe data integer dan float, bukan object.
   - Menggunakan IterativeImputer karena imputasi ini lebih baik dalam variasi data daripada median dan mean.
   - Menggunakan Robust Scaler karena meski outliers sudah diatasi, masih tetap terdapat outliers.
2. Binary : `Make`, `Type`, `Color`, dan `Region`.
   - Karena fitur-fitur ini memiliki lebih dari 5 unique, yang jika menggunakan One-Hot, dapat menyebabkan Overfitting.
3. Ordinal : `Options`
   - Karena value di dalamnya adalah _Full_ yang memiliki arti fitur lengkap, dan *Semi_Full* berarti tidak fitur lengkap namun lebih daripada *Standard*, maka fitur ini bersifat ordinal dan bukan nominal.
4. One-Hot : `Origin`, `Fuel_Type`, dan `Gear_Type`
   - Karena fitur-fitur ini memiliki fitur unique kurang dari 5.

```python
# Importing necessary libraries
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import RobustScaler, OrdinalEncoder, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from category_encoders import BinaryEncoder

# Membuat pipeline

# Preprocessing fitur-fitur numerikal
num_features = ['Year', 'Engine_Size', 'Mileage']
num_transformer = Pipeline(steps=[
    ('imputer', IterativeImputer()),
    ('scaler', RobustScaler())
])

# Preprocessing fitur ordinal - options
opt_ord_features = ['Options']
opt_ord_transformer = Pipeline(steps=[
    ('encoder', OrdinalEncoder(categories= [['Standard', 'Semi Full', 'Full']]))
])

# Preprocessing fitur-fitur kategorikal dengan Binary Encoder
binary_features = ['Make', 'Type', 'Color', 'Region']
binary_transformer = Pipeline(steps=[
    ('encoder', BinaryEncoder())
])

# Preprocessing fitur-fitur kategorikal lainnya
cat_features = ['Origin', 'Fuel_Type', 'Gear_Type']
cat_transformer = Pipeline(steps=[
    ('encoder', OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False))
])

# Mengkombinasikan transformer dengan ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features),
        ('ord_options', opt_ord_transformer, opt_ord_features),
        ('binary', binary_transformer, binary_features),
        ('cat', cat_transformer, cat_features)
    ], remainder='passthrough', verbose_feature_names_out=False)

preprocessor
```

### Model Benchmarking

**Assign Model**

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer, mean_absolute_error, mean_squared_error,mean_absolute_percentage_error, r2_score, median_absolute_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, BaggingRegressor, VotingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.naive_bayes import GaussianNB
from lightgbm import LGBMRegressor

# Define model names and list of models
model_name = ['Linear Regression', 'Ridge', 'Lasso', 'Decision Tree', 'Random Forest',
              'Gradient Boosting', 'SVR', 'Bagging', 'Voting', 'XGBoost', 'Naive Bayes', 'LightGBM']
list_model = [LinearRegression(), Ridge(), Lasso(), DecisionTreeRegressor(random_state=42),
              RandomForestRegressor(random_state=42), GradientBoostingRegressor(random_state=42),
              SVR(), BaggingRegressor(estimator=DecisionTreeRegressor(random_state=42)),
              VotingRegressor(estimators=[('lr', LinearRegression()), ('rf', RandomForestRegressor(random_state=42))]),
              XGBRegressor(random_state=42), GaussianNB(), LGBMRegressor(random_state=42)]
```

```python
# Create dictionary of models
models = dict(zip(model_name, list_model))

results = {}

scoring = {
    'MAE': make_scorer(mean_absolute_error, greater_is_better=False),
    'MedAE': make_scorer(median_absolute_error, greater_is_better=False),
    'RMSE': make_scorer(mean_squared_error, greater_is_better=False),
    'MAPE': make_scorer(mean_absolute_percentage_error, greater_is_better=False),
    'R2': make_scorer(r2_score)
}

# Loop for each model
for name, model in models.items():
    pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])
    cv_results = cross_validate(pipeline, xtrain, ytrain, cv=10, scoring=scoring, return_train_score=True)
    results[name] = cv_results
```

### CARA KERJA MODEL TERPILIH (XGBoost)

**XGBoost (eXtreme Gradient Boosting)**

XGBoost adalah implementasi yang dioptimalkan dari algoritma Gradient Boosting. Berikut adalah cara kerja XGBoost:

1. **Inisialisasi Model**: XGBoost dimulai dengan membuat prediksi awal, biasanya rata-rata dari target variabel.

2. **Pembangunan Trees**: XGBoost membangun decision trees secara berurutan. Setiap tree baru berfokus pada residual (kesalahan) dari tree sebelumnya.

3. **Feature Importance**: XGBoost menghitung skor penting untuk setiap fitur, memungkinkan pemilihan fitur yang lebih efisien.

4. **Regularisasi**: XGBoost menerapkan regularisasi L1 dan L2 untuk mencegah overfitting.

5. **Pruning**: XGBoost melakukan pruning level-wise, bukan leaf-wise, yang membantu mengurangi overfitting.

6. **Handling Missing Values**: XGBoost dapat menangani nilai yang hilang secara otomatis dengan mencari arah terbaik.

7. **Parallel Processing**: XGBoost mendukung komputasi paralel untuk membangun trees, meningkatkan kecepatan training.

8. **Cross-validation**: XGBoost menyediakan built-in cross-validation pada setiap iterasi.

9. **Early Stopping**: XGBoost dapat menghentikan training jika performa tidak meningkat setelah beberapa iterasi.

10. **Ensemble Learning**: Prediksi akhir adalah kombinasi dari semua trees yang dibangun.

XGBoost unggul dalam kecepatan dan performa, terutama untuk dataset terstruktur/tabular, membuatnya menjadi pilihan populer untuk banyak tugas machine learning.

### **Shapley Additive exPlanations (SHAP)**

```python
import shap # Import the shap module

# Convert categorical features in xtrain to numerical using one-hot encoding
xtrain = pd.get_dummies(xtrain)

# Fit the model to the transformed training data
# best_xgb_model.fit(xtrain, ytrain) # This line caused the error
best_model.fit(xtrain, ytrain) # Use the defined and fitted best_model instead

# Inisialisasi explainer SHAP
explainer = shap.TreeExplainer(best_model)

# Ensure xtest has the same columns as xtrain after one-hot encoding
xtest = pd.get_dummies(xtest)
missing_cols = set(xtrain.columns) - set(xtest.columns)
for c in missing_cols:
    xtest[c] = 0
xtest = xtest[xtrain.columns]

# Hitung nilai SHAP untuk data test
shap_values = explainer.shap_values(xtest)

# Get feature names after one-hot encoding
new_feature_names = xtest.columns.tolist()  # Extract feature names

# Visualisasi SHAP summary plot
shap.summary_plot(shap_values, xtest, feature_names=new_feature_names)

# Simpan visualisasi sebagai PNG dengan resolusi 300 dpi
plt.savefig('shap_summary_plot_high_res.png', dpi=300)

# Menampilkan visualisasi SHAP summary plot
plt.show()
```

### **Local Interpretable Model-agnostic (LIME)**

```python
import lime
import lime.lime_tabular

# Inisialisasi LIME explainer
explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=xtrain.values,
    feature_names=xtrain.columns,
    mode='regression',
    # Add discretize_continuous=False to handle zero variance features
    discretize_continuous=False
)

# Pilih instance yang akan dijelaskan (misalnya, instance pertama dalam data test)
instance_idx = 0
instance = xtest.iloc[instance_idx]

# Dapatkan penjelasan LIME
explanation = explainer.explain_instance(
    data_row=instance.values,  # Data instance yang akan dijelaskan
    predict_fn=best_model.predict,  # Fungsi untuk melakukan prediksi
    num_features=15  # Jumlah fitur yang akan ditampilkan dalam penjelasan
)

# Tampilkan penjelasan
explanation.show_in_notebook(show_table=True, show_all=True)

# Simpan penjelasan ke file png
explanation.save_to_file('lime.png')
```
